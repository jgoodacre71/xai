# Demo 03: PatchCore on MVTec AD Bottle

This notebook is the narrative layer for the PatchCore hero demo. It keeps the implementation in `src/xai_demo_suite/` and uses package code to build the static report artefacts.

## Story arc

1. Train a PatchCore-style memory bank from nominal bottle images only.
2. Score anomalous test images by patch distance to nominal examples.
3. Explain each result with a coarse anomaly map, top anomalous patch, nearest normal patches, and source images.
4. Probe a nearest-normal patch replacement as a didactic counterfactual.
5. Check the top patch against the MVTec ground-truth mask when a mask is available.

The four XAI questions are: evidence, provenance, counterfactual change, and stability. This notebook covers the first three; stability is reserved for a later robustness demo.

## Preconditions

Prepare the MVTec AD bottle data before running the notebook:

```bash
./.venv/bin/xai-demo-data fetch mvtec_ad --category bottle
./.venv/bin/xai-demo-data prepare mvtec_ad --category bottle
```

Raw data, processed manifests, cached memory banks, and generated reports are local artefacts and are ignored by git.

````python
from pathlib import Path

from xai_demo_suite.reports.patchcore_bottle import (
    PatchCoreBottleReportConfig,
    build_patchcore_bottle_report,
)
````

````python
config = PatchCoreBottleReportConfig(
    manifest_path=Path("data/processed/mvtec_ad/bottle/manifest.jsonl"),
    output_dir=Path("outputs/patchcore_bottle"),
    max_examples=3,
)
config
````

````python
report_path = build_patchcore_bottle_report(config)
report_path
````

## What to inspect

Open `outputs/patchcore_bottle/index.html` after running the previous cell. For each selected example, inspect:

- the input image and coarse anomaly overlay;
- the top anomalous patch;
- nearest normal patch crops and full source images;
- the distance summary table;
- the nearest-normal patch replacement preview;
- the ground-truth localisation check when a mask is available.

Also open `outputs/patchcore_bottle/demo_card.html` for the release-style summary.

## Caveats

The current local default uses deterministic colour and texture patch features. This makes the demo reproducible without downloading model weights, but it is not final pretrained multi-scale PatchCore. The mask overlap panel is a coarse top-patch verification check, not benchmark-grade pixel-level evaluation.
