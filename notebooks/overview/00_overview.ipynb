# XAI Demo Suite Overview

## Why this notebook exists
This repository now treats the notebooks as the active demo surface. The goal is to be able to open a notebook, inspect the code directly, run it cell by cell, and discuss the result without depending on a separate HTML report surface.

The non-SHAP notebooks are intentionally becoming larger because the code is meant to be visible and portable. Refactoring can come later once the notebook-first workflow is stable.
## Data source note
The active non-SHAP demo notebooks are self-contained synthetic walkthroughs by default: the images are generated inside the notebook rather than loaded from external dataset files.
