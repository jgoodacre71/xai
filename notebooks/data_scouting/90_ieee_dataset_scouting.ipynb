# Demo 90 — IEEE dataset scouting register

## Learning goals

This is not a modelling demo. It is a controlled register for deciding whether IEEE DataPort candidates are suitable, permitted, and worth integrating into a future XAI demo.

The goal is to avoid ad hoc data hunting: every candidate should have access, licence, citation, work-permission status, demo fit, and next action recorded before it becomes part of the project.


```python
# ruff: noqa: E501
from __future__ import annotations

from pathlib import Path

import yaml
from IPython.display import Markdown, display

DEMO = "90 IEEE dataset scouting register"
DATA_MODE = "candidate_register"
EXTERNAL_DATA_REQUIRED = False
PROJECT_ROOT = Path.cwd()
MANIFEST_PATH = PROJECT_ROOT / "data" / "ieee_candidates.yaml"
MANIFEST_EXISTS = MANIFEST_PATH.exists()
DATASET_SOURCE = "IEEE DataPort candidate metadata; no dataset selected yet"
LICENCE_NOTE = "Dataset-specific; verify before work use. IEEE DataPort datasets require attribution/citation."
MISSING_FILES = [] if MANIFEST_EXISTS else [str(MANIFEST_PATH)]
SEED = 20260515

print(f"DEMO: {DEMO}")
print(f"DATA_MODE: {DATA_MODE}")
print(f"EXTERNAL_DATA_REQUIRED: {EXTERNAL_DATA_REQUIRED}")
print(f"MANIFEST_PATH: {MANIFEST_PATH}")
print(f"MANIFEST_EXISTS: {MANIFEST_EXISTS}")
print(f"PROJECT_ROOT: {PROJECT_ROOT}")
print(f"DATASET_SOURCE: {DATASET_SOURCE}")
print(f"LICENCE_NOTE: {LICENCE_NOTE}")
print(f"MISSING_FILES: {MISSING_FILES}")
print(f"SEED: {SEED}")
```

## Why this register exists

IEEE DataPort can contain useful industrial, sensor, inspection, and fault datasets, but access and terms vary. This notebook keeps the project honest: a candidate is not part of the demo suite until it has a clear permission and relevance record.


```python
if MANIFEST_EXISTS:
    candidate_data = yaml.safe_load(MANIFEST_PATH.read_text(encoding="utf-8")) or {}
else:
    candidate_data = {"candidates": [], "schema": {}}

candidates = candidate_data.get("candidates", [])
schema = candidate_data.get("schema", {})
required_fields = schema.get("required_fields", [])
fit_categories = schema.get("fit_categories", {})

print(f"candidate_count: {len(candidates)}")
print(f"required_fields: {len(required_fields)}")
print(f"fit_categories: {', '.join(fit_categories.keys()) if fit_categories else 'none'}")
```

## Candidate schema

Each candidate should record slug, title, IEEE DataPort URL, DOI, access type, licence, citation, domain, task, data type, size, download status, work permission status, suitable demos, why it is interesting, risks, and next action.

Fit categories:

- `A`: excellent candidate; improves a core demo.
- `B`: useful backup.
- `C`: interesting but not urgent.
- `D`: poor fit or unclear terms.


```python
def render_candidate_table(candidates: list[dict[str, object]]) -> str:
    if not candidates:
        return "No IEEE candidates have been selected yet. Add candidates to `data/ieee_candidates.yaml` only after recording access, terms, and next action."
    lines = ["| Fit | Slug | Title | Access | Work permission | Suitable demos | Next action |", "| --- | --- | --- | --- | --- | --- | --- |"]
    for candidate in candidates:
        lines.append(
            "| "
            + " | ".join(
                [
                    str(candidate.get("fit", "unknown")),
                    str(candidate.get("slug", "")),
                    str(candidate.get("title", "")),
                    str(candidate.get("access_type", "unknown")),
                    str(candidate.get("work_permission_status", "verify before work use")),
                    ", ".join(candidate.get("suitable_demos", [])) if isinstance(candidate.get("suitable_demos", []), list) else str(candidate.get("suitable_demos", "")),
                    str(candidate.get("next_action", "record terms")),
                ]
            )
            + " |"
        )
    return "\n".join(lines)


display(Markdown(render_candidate_table(candidates)))
```

## What this does not prove

This register does not grant permission to use any dataset. It also does not imply that IEEE DataPort is better than existing open datasets. It is a controlled scouting workflow for future approved candidates.


```python
assert DATA_MODE == "candidate_register"
assert EXTERNAL_DATA_REQUIRED is False
assert MANIFEST_EXISTS
assert isinstance(candidates, list)
assert "access_type" in required_fields
assert "work_permission_status" in required_fields
print("Demo 90 self-check passed.")
```
